```python
import pandas as pd
```

Xử lý, làm sạch bộ dữ liệu Top 50 Animation Movies and TV Show và EDA để tìm ra các thông tin

```python
df1 = pd.read_csv("/content/Top 50 Animation Movies and TV Shows.csv")
df1.head(10)
```

```python
df1.info()
```

```python
df1.dropna(inplace= True)
df1['Minutes'] = df1['Minutes'].astype(str).str.replace(' min', '').astype(float)
df1['Votes']=df1['Votes'].astype(str).str.replace(',', '').astype(float)
```

```python
df1.head()
```

```python
df1.info()
```

```python
df1_exploded = df1.assign(genre=df1['genre'].str.split(', ')).explode('genre')
df1_exploded = df1_exploded[df1_exploded['genre'] != 'Animation']
```

```python
df1.head()
```

```python
average_duration_votes = df1_exploded.groupby('genre').agg({'Minutes': 'mean', 'Votes': 'mean'}).reset_index()
average_duration_votes
```

```python
average_ratings = df1_exploded.groupby('genre')['Rating'].mean().reset_index()
average_ratings
```

Bài tập 2: Hãy tiến hành xử lý, làm sạch bộ dữ liệu  Raw_credits và tìm ra các thông tin

```python
df2 = pd.read_csv("/content/raw_titles.csv")
df2.head(10)
```

```python
df2.info()
```

```python
df2['age_certification'].fillna('Unknown', inplace=True)
median_seasons = df2['seasons'].median()
df2['seasons'].fillna(median_seasons, inplace=True)
df2.info()
```

```python
df2.dropna(inplace= True)
df2.info()
```

```python
df2.head()
```

```python
average_imdb_score = df2['imdb_score'].mean()
print(f"Điểm số trung bình IMDb của tất cả các bộ phim: {average_imdb_score:.1f}")
```

```python
print(df2['genres'].head())
```

```python
def clean_genres(genres_str):
    genres_str = genres_str.strip("[]").replace("'", "")
    return ', '.join([genre.strip() for genre in genres_str.split(',')])

df2['genres'] = df2['genres'].apply(clean_genres)
df2.head()
```

```python
average_imdbscore_bygenres = df2.groupby('genres')['imdb_score'].mean().reset_index()
average_imdbscore_bygenres.columns = ['Genre', 'AverageIMDbScore']
average_imdbscore_bygenres
```

```python
df2['genres'] = df2['genres'].astype(str).str.split(', ')
df2_exploded = df2.explode('genres')

total_imdb_votes = df2['imdb_votes'].sum()
print(f"Tổng số IMDb votes của tất cả các bộ phim: {total_imdb_votes}")

imdb_votes_by_genres = df2_exploded.groupby('genres')['imdb_votes'].sum().reset_index()
imdb_votes_by_genres.columns = ['Genres', 'Total_IMDb_Votes']

print("\nSố lượng IMDb votes theo thể loại:")
imdb_votes_by_genres
```

```python
average_imdb_votes_and_score_by_year = df2.groupby('release_year')[['imdb_votes', 'imdb_score']].mean().reset_index()
average_imdb_votes_and_score_by_year
```
